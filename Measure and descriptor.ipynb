# Outliers

```python
#outliers with z score
```

```python

mean_x = df_['x'].mean()
mean_y = df_['y'].mean()
std_x = df_['x'].std()
def outliers(k):
    ol = []
    for i in range(len(df_)):
        diff = abs(df_['x'][i] - mean_x)/std_x
        if diff > k:
            ol.append(1)
        else:
            ol.append(0)
    df_['outliers'] = ol
    num_outliers = df_.query('outliers == 1').count()['x']
    print('number of outliers:{}'.format(num_outliers))
    sns.scatterplot(x='x',y='y',data=df_,hue='outliers')

```

```python
df_.query('outliers == 1').count()['x']
```

```python
outliers(1)   # one sigma
```

```python
outliers(2)   #two sigma  
```

```python
outliers(3)  #three sigma
```

```python
# outliers with MAD (Median absolute difference)
```

```python

x_median = df_['x'].median()
df_mad = abs(df_['x'] - x_median)
mad = df_mad.median()
ol = []
for i in range(len(df_['x'])):
    
    z_score = (0.6745 * (df_['x'][i] - x_median))/mad
    if z_score > 3 or z_score < -3:
        ol.append(1)
    else:
        ol.append(0)
df_['outliers_mad'] = ol
print('outliers count:',df_.query('outliers_mad == 1').count()['outliers_mad'])
sns.scatterplot(x='x', y='y', data=df_, hue='outliers_mad')
```
